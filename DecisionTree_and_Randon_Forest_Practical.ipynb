```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics as sm
```

```python
df=pd.read_csv(r"C:\Users\Sangeeth Kumar\ML_Code_Files\airline_passenger_satisfaction_data\train.csv")
```

```python
df.head(5)
```

```python
df.columns
```

```python
df.shape
```

```python
df.drop(["Unnamed: 0", "id"], axis = 1, inplace = True)
```

```python
df.isna().sum()* 100 / len(df)
```

```python
df.dropna(inplace = True)
```

```python

d = {"Business" : 3, "Eco Plus" : 2, "Eco" : 1}
df['Class'] = df['Class'].map(d)
df['Class'].unique()
```

```python
d = {"neutral or dissatisfied" : 0, "satisfied" : 1}
df['satisfaction'] = df['satisfaction'].map(d)
df['satisfaction'].unique()
```

```python
df.head()
```

```python
df.select_dtypes(include = "number").columns
```

```python
df = pd.get_dummies(df, columns = ['Gender', 'Customer Type', 'Type of Travel'])
df.head(2)
```

```python
Y = df['satisfaction']
X = df.loc[:, df.columns != "satisfaction"]
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, Y, test_size = 0.2, random_state = 77)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

__Decision Tree to predict Customer Satisfacation__

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
DT = DecisionTreeClassifier(class_weight = "balanced")
DT.fit(X_train, y_train)
```

```python
test_predictions = DT.predict(X_test)
train_predictions = DT.predict(X_train)
```

```python
test_predictions
```

```python
print(confusion_matrix(y_test, test_predictions))
print(classification_report(y_test, test_predictions))

print(confusion_matrix(y_train, train_predictions))
print(classification_report(y_train, train_predictions))
```

__Random Forest Classifier__

```python
RF = RandomForestClassifier(n_estimators = 99, class_weight= "balanced")
RF.fit(X_train, y_train)
```

```python
test_predictions = RF.predict(X_test)
train_predictions = RF.predict(X_train)
```

```python
print(confusion_matrix(y_test, test_predictions))
print(classification_report(y_test, test_predictions))

print(confusion_matrix(y_train, train_predictions))
print(classification_report(y_train, train_predictions))
```
